```python
from paddleocr import PaddleOCR
# Initialise the OCR model
reader = PaddleOCR(use_textline_orientation=True, lang='ch')
```

```python
import cv2
import re
import numpy as np
import pytesseract
import json
import os
import glob
from paddleocr import PaddleOCR

# Helper functions
def fix_common_confusions(text):
    """
        Simple fix for common OCR errors
    """
    text = text.replace('I', '1').replace('O', '0')
    return text

def clean_text(text):
    """
        Removes special characters but keeps Chinese characters
        Range \u4e00-\u9fa5 covers common Chinese characters
    """
    text = text.upper()
    # Keep A-Z, 0-9, and Chinese characters
    return re.sub(r'[^A-Z0-9\u4e00-\u9fa5]', '', text)
    
def validate_plate_format(text):
    """
        Checks if text matches Taiwan, Malaysia or China formats.
        Returns: (True/False, Cleaned_Text, Detected_Region)
    """
    clean = clean_text(text)
    
    # China format
    # Structure: 1 Chinese char + 1 letter + 5-6 alphanumeric
    # Green plates have 6 digits at the end
    china_pattern = r'^[\u4e00-\u9fa5][A-Z][A-Z0-9]{5,6}$'
    if re.match(china_pattern, clean):
        return True, clean, "CHINA"

    # Malaysia format
    # Structure: 1-3 letters + 1-4 digits + optional 1 letter suffix
    malaysia_pattern = r'^[A-Z]{1,3}[0-9]{1,4}[A-Z]?$'
    
    # Taiwan format (AOLP)
    # Examples: AB-1234, 1234-AB, ABC-5678
    taiwan_patterns = [
        r'^[A-Z]{2,3}[0-9]{3,4}$', 
        r'^[0-9]{4}[A-Z]{2}$',      
        r'^[A-Z]{2}[0-9]{4}$',      
        r'^[0-9]{2,4}[A-Z]{1,2}$'   
    ]

    # Check Malaysia
    if re.match(malaysia_pattern, clean):
        # Additional check: Malaysia rarely has >8 chars
        if len(clean) <= 8:
            return True, clean, "MALAYSIA"

    # Check Taiwan
    for pat in taiwan_patterns:
        if re.match(pat, clean):
            return True, clean, "TAIWAN"

    return False, clean, "UNKNOWN"

def levenshtein_distance(s1, s2):
    """
        Calculates the edit distance between two strings
    """
    if len(s1) < len(s2):
        return levenshtein_distance(s2, s1)

    if len(s2) == 0:
        return len(s1)

    previous_row = range(len(s2) + 1)
    for i, c1 in enumerate(s1):
        current_row = [i + 1]
        for j, c2 in enumerate(s2):
            insertions = previous_row[j + 1] + 1
            deletions = current_row[j] + 1
            substitutions = previous_row[j] + (c1 != c2)
            current_row.append(min(insertions, deletions, substitutions))
        previous_row = current_row
    
    return previous_row[-1]

def get_ground_truth(subset_name, image_filename):
    """
        Reads the groundtruth recogntion text file from the original AOLP dataset paths
        Path format: /kaggle/input/aolp-dataset/AOLP/{Subset}/groundtruth_recognition/{filename}.txt
    """
    # Remove extension from image ('1.jpg' to '1')
    base_name = os.path.splitext(image_filename)[0]
    txt_filename = base_name + ".txt"
    gt_path = f"/kaggle/input/aolp-dataset/AOLP/{subset_name}/groundtruth_recognition/{txt_filename}"    
    
    if os.path.exists(gt_path):
        with open(gt_path, 'r') as f:
            # Read lines, strip whitespace, and join (e.g., "D\nF\n7" -> "DF7")
            lines = f.readlines()
            ground_truth = "".join([line.strip() for line in lines])
            return ground_truth
    return None

def ocr_voting(cropped_plate_img):
    if len(cropped_plate_img.shape) == 2:
        cropped_plate_img = cv2.cvtColor(cropped_plate_img, cv2.COLOR_GRAY2BGR)

    # It is better to have plate height to be at least 90 pixels
    # so characters are around 30-40 pixels tall (ideal for OCR)
    h, w = cropped_plate_img.shape[:2]
    target_height = 90.0
    if h < target_height:
        scale = target_height / float(h)
        new_w = int(w * scale)
        new_h = int(target_height)
        cropped_plate_img = cv2.resize(cropped_plate_img, (new_w, new_h), interpolation=cv2.INTER_CUBIC)
        kernel = np.array([[0, -1, 0], [-1, 5,-1], [0, -1, 0]])
        cropped_plate_img = cv2.filter2D(cropped_plate_img, -1, kernel)

    output_dir = "paddle_output"
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    paddle_text = ""
    paddle_conf = 0.0
    
    try:
        # Pass the image to Paddle OCR to predict
        result = reader.predict(cropped_plate_img) 
        for res in result:
            # Save PaddleOCR result to JSON and image, then load JSON for text and confidence
            res.save_to_img(output_dir)
            res.save_to_json(output_dir)  
            list_of_files = glob.glob(os.path.join(output_dir, '*.json'))
            if list_of_files:
                latest_file = max(list_of_files, key=os.path.getctime)
                with open(latest_file, 'r') as f:
                    data = json.load(f) 
                if 'rec_texts' in data and data['rec_texts']:
                    paddle_text = "".join(data['rec_texts'])
                if 'rec_scores' in data and data['rec_scores']:
                    scores = data['rec_scores']
                    paddle_conf = sum(scores) / len(scores)
                os.remove(latest_file)
            break
    except Exception:
        pass

    paddle_text = fix_common_confusions(paddle_text)
    is_valid_paddle, clean_paddle, _ = validate_plate_format(paddle_text)

    # Decision 1: If Paddle OCR sees a valid China/Malaysia/Taiwan plate with high conf
    if is_valid_paddle and paddle_conf > 0.80:
        return clean_paddle, paddle_conf, f"PaddleOCR (High Conf)"

    # Run Tesseract (Fallback) 
    # The Tesseract is configure to return alphanumeric and Chinse province characters on number plates only
    config = r'--oem 3 --psm 7 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789京沪粤津渝苏浙鲁皖闽晋川贵湘赣辽黑吉'
    try:
        tess_text = pytesseract.image_to_string(cropped_plate_img, config=config)
        tess_text = fix_common_confusions(tess_text)
        is_valid_tess, clean_tess, _ = validate_plate_format(tess_text)
    except:
        clean_tess = ""
        is_valid_tess = False

    # Voting logic
    # Priority A: If Paddle found a valid format (China/Malaysia/Taiwan), trust it.
    if is_valid_paddle:
        return clean_paddle, paddle_conf, "PaddleOCR (Validated)"

    # Priority B: If Tesseract found a valid format
    if is_valid_tess:
        return clean_tess, 0.70, "Tesseract (Validated)"

    # Priority C: Fallback to best guess (whichever is longer/more complete)
    if len(clean_paddle) >= len(clean_tess):
        return clean_paddle, paddle_conf, "PaddleOCR (Best Effort)"
    else:
        return clean_tess, 0.50, "Tesseract (Best Effort)"



root_path = "/kaggle/input/phase-3-4-results/Phase3_4_Results"
print(f"Checking Root Path: {root_path}")

total_distance = 0
total_chars = 0
processed_count = 0

print(f"{'FILENAME':<20} | {'PREDICTED':<10} | {'GROUND TRUTH':<12} | {'CER':<5} | {'METHOD':<20} | {'PADDLE CONF'}")
print("-" * 80)

for subdir, dirs, files in os.walk(root_path):
    for filename in files:
        if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            
            file_path = os.path.join(subdir, filename)
            
            subset_name = os.path.basename(subdir) 
            
            img = cv2.imread(file_path)
            if img is None: continue

            try:
                # Get prediction
                pred_text, conf, method = ocr_voting(img)
                
                # Get ground truth
                gt_text = get_ground_truth(subset_name, filename)
                
                if gt_text:
                    # Calculate CER 
                    dist = levenshtein_distance(pred_text, gt_text)
                    length = len(gt_text)
                    
                    # Avoid division by zero
                    if length > 0:
                        cer = dist / length
                    else:
                        cer = 1.0 if dist > 0 else 0.0

                    # Update globals for average CER
                    total_distance += dist
                    total_chars += length
                    processed_count += 1
                    
                    print(f"{filename:<20} | {pred_text:<10} | {gt_text:<12} | {cer:.2f}  | {method:<20} | {conf:.2f}")
                else:
                    print(f"{filename:<20} | {pred_text:<10} | {'NOT FOUND':<12} | {'N/A':<5} | {method:<20} | {conf:.2f}")

            except Exception as e:
                print(f"Error processing {filename}: {e}")

print("-" * 80)
if total_chars > 0:
    avg_cer = total_distance / total_chars
    accuracy = 1.0 - avg_cer
    print(f"Total Processed: {processed_count}")
    print(f"Total Characters: {total_chars}")
    print(f"Total Errors (Edit Distance): {total_distance}")
    print(f"AVERAGE CER: {avg_cer:.4f} ({avg_cer*100:.2f}%)")
    print(f"Approx Accuracy: {accuracy*100:.2f}%")
else:
    print("No ground truth data found or processed.")
```
